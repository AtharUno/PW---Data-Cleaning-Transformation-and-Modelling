Question 1 : What is Data Loading in Power BI and why is it considered the first step of analysis?

-Ans: Data Loading in Power BI is the process of connecting to external data sources (such as CSV files, Excel workbooks, SQL databases, or Web APIs) and importing or querying that raw data into Power BI Desktop.

Why it is the first step of analysis:

Foundation of Insights: Analysis cannot begin without underlying data. Importing the data creates the base layer upon which data cleaning, modeling, and visualization take place.

Schema & Type Recognition: When data is loaded, Power BI reads metadata, detects data types (e.g., Text, Integer, Date), and identifies table structures automatically.

Storage Engine Initialization: It populates Power BI’s internal VertiPaq engine (in Import Mode), compressing the data in memory for high-speed analytical queries.

Question 2 : Explain the difference between “Load” and “Transform Data” in Power BI.

Answer: Load Data:

-Primary Action:Imports raw data directly into the Power BI data model without editing.
-When to Use:	When the data is already clean, properly formatted, and ready for analysis.
-Performance: Faster initial import, but may clutter the model with dirty/unnecessary data.

Transform Data
-Primary Action:Opens the Power Query Editor to modify, clean, and reshape the dataset first.
-When to Use:When the dataset contains nulls, duplicates, inconsistent formats, or extra columns.
-Performance:Adds an extra step initially, but optimizes model performance by removing garbage data.

Question 3 : What is a Fact Table and a Dimension Table? Give examples from the dataset.

-Ans:

Fact Table:
Definition: A central table in a data model that contains quantitative/numerical measurements, metrics, and transaction records. It typically has many rows and contains foreign keys referencing dimension tables.

Example:Confirmed_Cases, Deaths, Recovered_Cases, and Vaccinated_Count, along with foreign keys like Date_ID and Location_ID.

Dimension Table:

Definition: A table that stores descriptive attributes/context used to slice, filter, and group the numerical measures in the fact table.

Example:

Dim_Location (containing Country, State/Province, Latitude, Longitude, Population).

Dim_Date (containing Date, Year, Quarter, Month).



Question 4 : Why is Star Schema preferred over Snowflake Schema in Power BI?

-Ans: A Star Schema connects dimension tables directly to a central fact table (1-level deep), whereas a Snowflake Schema normalizes dimensions into sub-tables (multi-level deep).
Why Star Schema is preferred in Power BI:
Better Query Performance: Power BI's is specifically optimized for Star Schemas. It minimizes complex JOIN operations, making visuals render much faster.
Simpler DAX Calculations: Writing Data Analysis Expressions (DAX) is simpler and less error-prone when dimension fields sit in a single context table rather than across multiple normalized tables.
Easier Navigation: The Model View is cleaner and easier for analysts and end-users to understand.
Optimal Relationship Direction: Supports clear 1-to-Many single-direction relationships natively.

Question 5: Identify and remove duplicate records based on Date, Country, and State.  

-Ans:
Method: Using Power Query Editor:
1. Open Power Query Editor by clicking Transform Data on the Home tab.
2. Select the dataset table.
3. Multi-select the three columns by holding Ctrl and clicking: Date, Country, and State.
4. Right-click on any selected header and choose Remove Duplicates.


Question 6: Identify and replace null values in Vaccination_Status.

-Ans:
Method:
1. In Power Query Editor, right-click the header of the Vaccination_Status column.
2. Select Replace Values...
3. In Value To Find, type null (or leave it blank).
4. In Replace With, type "Not Specified" (or "Unknown").
5. Click OK.

Question 7: Create a new column to calculate Recovery Rate.

-Ans:
Calculated Column via DAX:

1. Go to Data View, right-click the table, and select New Column.
2. Enter the following DAX formula:

= if [Confirmed_Cases] <> null and [Confirmed_Cases] > 0
  then ([Recovered_Cases] / [Confirmed_Cases]) * 100
  else 0

Question 8: Create a summarized table showing total confirmed cases by Country.

-Ans:
In Power Query, select the Country column.
Go to the Transform tab and click Group By.
Set New column name = Total Confirmed Cases.
Set Operation = Sum.
Set Column = Confirmed_Cases.
